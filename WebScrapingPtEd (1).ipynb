```python
import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = "https://www.cancer.gov/types/head-and-neck/hp/adult/oropharyngeal-treatment-pdq"

# Send a GET request
response = requests.get(url)

# Function to print all section tags following a specific header (e.g., h2)
def print_following_sections(soup, header_text):
    # Find the specific header that matches the header_text
    header = soup.find(lambda tag: tag.name == 'h2' and header_text in tag.get_text())

    if header:
        # Iterate through next siblings and print all sections
        for sibling in header.find_all_next():
            if sibling.name == 'section':
                print(sibling.get_text(strip=True))
                print("\n---\n")  # Separator for readability

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Specify the text within the header from which to start printing sections
    start_header_text = "General Information about Oropharyngeal Cancer"  # Change this to match the header's text
    print_following_sections(soup, start_header_text)
else:
    print("Failed to retrieve the webpage, status code:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = "https://www.cancer.gov/types/head-and-neck/hp/adult/oropharyngeal-treatment-pdq"

# Send a GET request
response = requests.get(url)

# Function to extract all text following each h2 within a section
def extract_content(section):
    content = []
    # Find all h2 elements within the section
    headers = section.find_all('h2')
    for header in headers:
        # Append header text to content
        content.append(header.get_text(strip=True) + '\n')
        # Initialize next sibling of the header
        next_sibling = header.find_next_sibling()
        # Continue until no more siblings or another h2 is found
        while next_sibling and next_sibling.name != 'h2':
            if next_sibling.name == 'p' or next_sibling.name == 'div':
                content.append(next_sibling.get_text(strip=True) + '\n')
            elif next_sibling.name == 'ul':
                # Extract text from each list item
                for li in next_sibling.find_all('li'):
                    content.append(li.get_text(strip=True) + '\n')
            next_sibling = next_sibling.find_next_sibling()
    return content

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all sections or divs that are likely to contain the structured content
    sections = soup.find_all('div', class_="accordion")

    # Prepare to write to file
    with open('PDQ_physician.txt', 'w', encoding='utf-8') as file:
        # Loop through each section and extract content
        for section in sections:
            extracted_content = extract_content(section)
            if extracted_content:
                for item in extracted_content:
                    file.write(item)
                file.write("---\n")  # Separator for readability
            else:
                file.write("No content found under this section.\n")
else:
    print("Failed to retrieve the webpage, status code:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

# URL to scrape
url = "https://about-cancer.cancerresearchuk.org/about-cancer/mouth-cancer/getting-diagnosed/tests-mouth-oropharyngeal-cancer"

# Send a GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the main content anchor and its nearest container with actual content
    main_content_anchor = soup.find('a', id='main-content')
    if main_content_anchor:
        # Assuming the content starts right after the <a id="main-content"> within the next div.region-content
        region_content = main_content_anchor.find_next_sibling('div', class_='region-content')
        if region_content:
            # Extract all relevant textual content
            elements = region_content.find_all(['h1', 'h2', 'h3', 'p', 'ul'])
            for element in elements:
                if element.name == 'ul':
                    for li in element.find_all('li'):
                        print(li.get_text(strip=True))
                else:
                    print(element.get_text(strip=True))
            print("\n---\n")  # Separator for readability
        else:
            print("Content region not found.")
    else:
        print("Main content anchor not found.")
else:
    print("Failed to retrieve the webpage, status code:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup
import os

# List of URLs to scrape
cruk_urls = [
    "https://about-cancer.cancerresearchuk.org/about-cancer/mouth-cancer/about",
    "https://about-cancer.cancerresearchuk.org/about-cancer/mouth-cancer/getting-diagnosed/tests-mouth-oropharyngeal-cancer",
    "https://about-cancer.cancerresearchuk.org/about-cancer/mouth-cancer/living-with/eating",
    "https://about-cancer.cancerresearchuk.org/about-cancer/mouth-cancer/treatment/surgery/help-you-breathe-or-eat",
    "https://about-cancer.cancerresearchuk.org/about-cancer/mouth-cancer/treatment/chemoradiotherapy/side-effects",
    "https://about-cancer.cancerresearchuk.org/about-cancer/mouth-cancer/treatment/radiotherapy/planning-radiotherapy",
    "https://about-cancer.cancerresearchuk.org/about-cancer/mouth-cancer/treatment/radiotherapy/external-radiotherapy",
    "https://about-cancer.cancerresearchuk.org/about-cancer/mouth-cancer/treatment/follow-up"
]

# Create session object to persist certain parameters across requests
session = requests.Session()

# Ensure the output directory exists
output_dir = 'output'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop through each URL
for index, url in enumerate(cruk_urls):
    # Send a GET request
    response = session.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the main content anchor and its nearest container with actual content
        main_content_anchor = soup.find('a', id='main-content')
        if main_content_anchor:
            # Assuming the content starts right after the <a id="main-content">
            region_content = main_content_anchor.find_next('div', class_='region-content')
            if region_content:
                # Define file name based on URL index
                file_path = os.path.join(output_dir, f'content_{index+1}.txt')
                with open(file_path, 'w', encoding='utf-8') as file:
                    # Extract all relevant textual content
                    elements = region_content.find_all(['h1', 'h2', 'h3', 'p', 'ul'])
                    for element in elements:
                        if element.name == 'ul':
                            for li in element.find_all('li'):
                                file.write(li.get_text(strip=True) + '\n')
                        else:
                            file.write(element.get_text(strip=True) + '\n')
                print(f"Content extracted and saved to {file_path}")
                print("---\n")  # Separator for readability
            else:
                print("Content region not found for URL:", url)
        else:
            print("Main content anchor not found for URL:", url)
    else:
        print(f"Failed to retrieve {url}, status code: {response.status_code}")
```
